```python
# Esta seção introduz os conceitos de raciocínio lógico, incluindo a lógica dedutiva e indutiva [6, 11].
# No cenário-problema da seção, a tarefa é usar esses tipos de raciocínio para inferir a probabilidade de um negócio ser fechado com um cliente [7, 9].
# A lógica é a base para a construção de algoritmos e o desenvolvimento de programas de computador [3, 6].

print("--- Análise de Fechamento de Negócio (Baseado na Lógica Indutiva e Dedutiva) ---")

# Premissas baseadas em raciocínio indutivo:
# No problema da seção, o raciocínio indutivo parte de observações particulares (dezenas de testes, históricos de aquisição)
# para chegar a conclusões gerais, mesmo que um contraexemplo possa invalidá-las [12-14].

# A partir da frase: "Realizamos dezenas de testes com o seu software e em todos ele foi capaz de chegar à melhor solução para nosso problema." [15]
# Inferimos indutivamente que:
software_resolve_problema_cliente = True
print(f"1. Conclusão (Indutiva): O software da empresa resolve o problema do cliente? {software_resolve_problema_cliente} [16]")

# A partir da frase: "Nos últimos anos, os softwares que resolveram nossos problemas foram adquiridos para a melhoria de nossos processos." [16]
# Inferimos indutivamente que:
cliente_sempre_adquire_softwares_que_resolvem_problemas = True
print(f"2. Conclusão (Indutiva): O cliente sempre adquire softwares que resolvem seus problemas? {cliente_sempre_adquire_softwares_que_resolvem_problemas} [16]")

# Aplicação do Raciocínio Dedutivo:
# O raciocínio dedutivo parte de premissas gerais para chegar a uma conclusão particular e inevitável, se as premissas forem verdadeiras [14, 17].
# As duas conclusões indutivas acima servem como premissas para a dedução final [10].

print("\n--- Aplicando Raciocínio Dedutivo para a Conclusão Final ---")

# Utilizamos uma estrutura condicional (if-else), que é um componente fundamental na construção de algoritmos [18],
# para simular a dedução lógica.
if software_resolve_problema_cliente and cliente_sempre_adquire_softwares_que_resolvem_problemas:
    venda_sera_fechada = True
    print("\n**Perspectiva: Positiva para o fechamento do negócio!**")
    print("Conclusão (Dedutiva): O software de sua empresa será adquirido. [10]")
else:
    venda_sera_fechada = False
    print("\n**Perspectiva: O fechamento do negócio não é garantido.**")
    print("Lembre-se: As conclusões do raciocínio indutivo são válidas até que ocorra um contraexemplo. [10, 13]")

print(f"\nStatus final inferido da venda: {venda_sera_fechada}")

# Este exemplo simples demonstra como os conceitos de lógica, raciocínio indutivo e dedutivo,
# introduzidos na Unidade 1, Seção 1, são a base para o pensamento computacional e a construção de algoritmos
# que tomam decisões [3, 6, 19].
```
